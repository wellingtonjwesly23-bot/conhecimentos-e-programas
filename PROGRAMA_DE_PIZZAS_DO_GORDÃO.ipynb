```python
# A. Mensagem de boas-vindas e Menu
print("Bem-vindo ao GORDÃO DA PIZZA!")
print("Bem-vindo Wellington Wesley!")

# Definindo sabores
SABORES_SAL = ["Calabresa", "Mussarela", "Frango com Catupiry"]
SABORES_DOCE = ["Chocolate", "Romeu e Julieta", "Doce de Leite"]

# Bolando o cardapio Adicionei 3 sabores para ficar mais completo
def exibir_cardapio():
    print("-----|--------------------------------- MENU --------------------------------| ")
    print("-----| Tamanho P: Pizza Salgada (PS) - R$ 30.00 | Pizza Doce (PD) - R$ 34.00 |-----")
    print("-----| Tamanho M: Pizza Salgada (PS) - R$ 45.00 | Pizza Doce (PD) - R$ 48.00 |-----")
    print("-----| Tamanho G: Pizza Salgada (PS) - R$ 60.00 | Pizza Doce (PD) - R$ 66.00 |-----")
    print("-----------------------------------------------------------------------------------")
    print("\nSabores Salgados:")
    for sabor in SABORES_SAL:
        print(f"- {sabor}")
    print("\nSabores Doces:")
    for sabor in SABORES_DOCE:
        print(f"- {sabor}")
    print("------------")


exibir_cardapio()

total_pedido = 0  # Acumulador para o total do pedido

while True:  # Loop para pedir mais coisas
    # Input do sabor (PS/PD) e validação
    while True:
        tipo_pizza = input("Digite o tipo de pizza (PS para Salgada, PD para Doce): ").upper()
        if tipo_pizza in ["PS", "PD"]:
            break
        else:
            print("Tipo de pizza inválido. Tente novamente.")

    # Input do tamanho (P/M/G) e validação
    while True:
        tamanho = input("Digite o tamanho da pizza (P, M, G): ").upper()
        if tamanho in ["P", "M", "G"]:
            break
        else:
            print("Tamanho inválido. Tente novamente.")

    # Seleção do sabor
    if tipo_pizza == "PS":
        while True:
            sabor = input(f"Escolha um sabor salgado ({', '.join(SABORES_SAL)}): ").capitalize()
            if sabor in SABORES_SAL:
                break
            else:
                print("Sabor inválido. Escolha um sabor da lista.")
    elif tipo_pizza == "PD":
        while True:
            sabor = input(f"Escolha um sabor doce ({', '.join(SABORES_DOCE)}): ").capitalize()
            if sabor in SABORES_DOCE:
                break
            else:
                print("Sabor inválido. Escolha um sabor da lista.")


    # Combinações de sabor e tamanho e cálculo do preço (modelo aninhado)
    if tipo_pizza == "PS":
        if tamanho == "P":
            preco = 30.0
        elif tamanho == "M":
            preco = 45.0
        elif tamanho == "G":
            preco = 60.0
    elif tipo_pizza == "PD":
        if tamanho == "P":
            preco = 34.0
        elif tamanho == "M":
            preco = 48.0
        elif tamanho == "G":
            preco = 66.0

    print(f"Adicionado: Pizza {tipo_pizza} - Tamanho {tamanho} - Sabor {sabor} - R$ {preco:.2f}")
    total_pedido += preco  # Acumula o valor do pedido

    # Pergunta se deseja pedir mais alguma coisa
    while True:
        continuar = input("Deseja pedir mais alguma coisa? (sim/nao): ").lower()
        if continuar in ["sim", "nao"]:
            break
        else:
            print("Resposta inválida. Digite 'sim' ou 'nao'.")

    if continuar == "nao":
        break

# Finaliza o prog
print("\n--- Resumo do Pedido ---")
print(f"Total a pagar: R$ {total_pedido:.2f}")
print("-------------------------")
```
